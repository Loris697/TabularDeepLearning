```python
aggregated_dataset = metadataset.groupby(["dataset_name", "alg_name"]).agg({
    "Accuracy__test" : "max",
    "F1__test" : "max"
})
aggregated_dataset = aggregated_dataset.reset_index().merge(metafeatures_datasets, on="dataset_name", how="left")
aggregated_dataset
```

```python
#Dropping the dataset where the Linear Model has a performance grater that .95
# Step 1: Identify the datasets where 'LinearModel' has 'Accuracy__test' > 0.95
high_perf_datasets = aggregated_dataset.query("(alg_name == 'LinearModel' or alg_name == 'DecisionTree') and Accuracy__test > 0.95")['dataset_name'].unique()

# Step 2: Filter out all rows from these datasets using a query that negates the inclusion of 'high_perf_datasets'
aggregated_dataset_difficult = aggregated_dataset.query("dataset_name not in @high_perf_datasets")
```

```python
# Reset index so 'dataset_name' and 'alg_name' are back as columns
grouped = aggregated_dataset_difficult.reset_index(drop=True)

# Find the max 'Accuracy_Test' within each 'dataset_name'
max_accuracy_per_dataset = grouped.groupby('dataset_name')['Accuracy__test'].transform("max")

# Filter rows where 'Accuracy_Test' matches the max 'Accuracy_Test' within each 'dataset_name'
best_algs = grouped[grouped['Accuracy__test'] == max_accuracy_per_dataset]
```

```python
merged_df = pd.merge(best_algs, classDF, on='alg_name')
merged_df["f__pymfe.general.nr_inst"] = np.log(merged_df["f__pymfe.general.nr_inst"])

sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
scatter = sns.scatterplot(x='f__pymfe.general.nr_inst',
                          y='f__pymfe.general.nr_cat', 
                          hue='Group', 
                          data=merged_df, 
                          palette='coolwarm', 
                          s=100,
                          alpha=0.5  # adjusted alpha for transparency
                         )
plt.title('Scatter Plot of nr_inst vs nr_cat Colored by Group')
plt.xlabel('Number of Instances (nr_inst)')
plt.ylabel('Number of Categories (nr_cat)')

# Setting the limits for x and y axes
#plt.xlim(-1000, 10000)
#plt.ylim(-1, 20)

plt.legend(title='Group', bbox_to_anchor=(1.05, 1), loc=2)
plt.show()
```
