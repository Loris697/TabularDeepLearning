## Splitting the Analysis

```python
small_datasets = metafeatures_datasets[metafeatures_datasets["f__pymfe.general.nr_inst"] < 1000 ].sort_values("f__pymfe.general.nr_inst")
datasets = metafeatures_datasets[metafeatures_datasets["f__pymfe.general.nr_inst"] >= 1000 ].reset_index()
small_datasets.head(15)
```

```python
datasets
```

```python
grouped_dataset = metadataset.groupby(["dataset_name", "alg_name"]).agg({
    "Accuracy__test" : "max",
    "F1__test" : "max"
})
```

```python
aggregated_dataset = grouped_dataset.reset_index().merge( datasets, on = "dataset_name", how = "inner")
aggregated_dataset = aggregated_dataset.merge(classDF, on = "alg_name")
```

```python
aggregated_dataset_small = grouped_dataset.reset_index().merge( small_datasets, on = "dataset_name", how = "inner")
aggregated_dataset_small = aggregated_dataset_small.merge(classDF, on = "alg_name")
```

## Finding Best Algorithms

```python
# Reset index so 'dataset_name' and 'alg_name' are back as columns
grouped = aggregated_dataset.reset_index(drop=True)
grouped_small = aggregated_dataset_small.reset_index(drop=True)

# Find the max 'Accuracy_Test' within each 'dataset_name'
max_accuracy_per_dataset = grouped.groupby('dataset_name')['Accuracy__test'].transform("max")
max_accuracy_per_dataset_small = grouped_small.groupby('dataset_name')['Accuracy__test'].transform("max")

# Filter rows where 'Accuracy_Test' matches the max 'Accuracy_Test' within each 'dataset_name'
best_algs = grouped[grouped['Accuracy__test'] == max_accuracy_per_dataset]
best_algs_small = grouped_small[grouped_small['Accuracy__test'] == max_accuracy_per_dataset_small]

#Taking just the datasets where there are not ties
tie_dataset = list(best_algs.groupby("dataset_name").nunique().query("Group > 1").index)
tie_dataset_small = list(best_algs_small.groupby("dataset_name").nunique().query("Group > 1").index)

best_algs = best_algs.query(" dataset_name not in @tie_dataset ")
best_algs_small = best_algs_small.query(" dataset_name not in @tie_dataset_small ")
```

```python
group_performance = aggregated_dataset.groupby(["dataset_name", "Group"]).agg({
                                                        "Accuracy__test" : "max",
                                                        "F1__test" : "max",
                                                        "f__pymfe.general.nr_inst" : "first"
                                                }).reset_index()
max_per_dataset = group_performance.groupby("dataset_name")["Accuracy__test"].max().reset_index().rename(columns = {"Accuracy__test" : "Accuracy_max"})
group_performance = group_performance.merge(max_per_dataset, on = "dataset_name")
group_performance["gap"] = group_performance.Accuracy__test - group_performance.Accuracy_max
```
